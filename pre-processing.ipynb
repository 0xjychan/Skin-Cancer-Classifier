```python
import pandas as pd
import numpy as np
import os
import shutil
```

```python
# Load training data from csv
train_data = pd.read_csv('train.csv')
train_data.head()
```

```python
# Move image data to train folder according to its target value
source_dir = 'temp'
img_dict = train_data[['image_name','target']].set_index('image_name').to_dict()['target']
```

```python
for file in os.listdir(source_dir):
    label = img_dict[file[:-4]]
    if label==0:
        shutil.move(os.path.join(source_dir, file),'train/~benign/'+file)
    else:
        shutil.move(os.path.join(source_dir, file),'train/~maglinant/'+file)
```
